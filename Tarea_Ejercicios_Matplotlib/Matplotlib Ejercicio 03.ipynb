# Ejercicios de matplotlib 

Matplotlib puede ser difícil de entender a la primera sin embargo es una librería muy útil para la visualización pueden apoyarse con la hoja de soluciones 



## Datos

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

importar matplotlib y agregar la linea para poder visualizar las gráficas en la consola de jupyter 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Ejercicio 1

- Crear un objeto usando plt.figure()
- Agregar un  eje (axis) a la figura 
- Graficar los ejes x,y  
- Agregar las etiquetas correspondientes 


```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("Eje X")
axes.set_ylabel("Eje Y")
```

## Ejercicio 3

Crear un objeto  y agregar dos ejes (axis) , en las siguientes locaciones [0,0,1,1] y [0.2,0.5,0.2,0.2]  


```python
fig2 = plt.figure()
axs = fig2.add_axes([0,0,1,1])
axs2 = fig2.add_axes([0.2,0.5,0.2,0.2])
```

- Graficar x,y en ambos ejes y llamar a la figura  

```python
axs.plot(x,y)
axs2.plot(x,y)
fig2
```

## Ejercicio 2

Crear un objeto  y agregar dos ejes (axis) , en las siguientes locaciones [0,0,1,1] y [0.2,0.5,.4,.4]  


```python
fig3 = plt.figure()
ax = fig3.add_axes([0,0,1,1])
ax2 = fig3.add_axes([0.2,0.5,0.4,0.4])
```

Graficar x,y en el eje de menor tamaño  y x,z en eje principal, poner particular atención en los límites de las figuras 

```python
ax.plot(x,z)
ax2.plot(x,y)
fig3
```

## Ejercicio 4

Usar subplots de una fila dos columnas 

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
```

Graficar (x,y) y (x,z) en los ejes (axis) con el grosor de 3 y el estilo adecuado 

```python
axes[0].plot(x,y,color="blue",lw=3,linestyle="--")
axes[1].plot(x,z,color="red",lw=3,linestyle="-")
fig
```
